Calculating the Average and Median

```python
so2 = data_new['a_so2_mean_ln']
so2_mean= so2.mean()
print('Mean: ',so2_mean)
so2_median = so2.median()
print('Median: ',so2_median)
```

Sampling

```python
np.random.seed(1)
sample = np.random.choice(so2,100)
sample_mean = sample.mean()
sample_median = np.median(sample)
print('Sample Mean: ',sample_mean)
print('Sample Median: ',sample_median)
```

Compare population and Sampling distribution

```python
sampledf = pd.DataFrame(sample)
fig, (ax1,ax2) = plt.subplots(2,1, sharex=True,)
ax1.hist(so2,bins=100)
ax2.hist(sampledf,bins=30)
ax2.set_xlabel('SO2')
ax1.set_ylabel('Frequency')
ax2.set_ylabel('Frequency')
plt.show()
```
