```python
data= pd.read_csv('/content/data_air.csv')
data
```

```python
data_new=data.insert(1,'StName',data['state_name'])
data_new = data.drop(['state_name','countyname','state','stfips','fips'],axis=1)
data_new.head()
```

```python
co = data_new['a_co_mean_ln']
```

```python
plt.boxplot(co)
plt.show()
```

```python
data_new.boxplot(column='a_co_mean_ln',by='StName',figsize=(13,8),grid=False,rot=45)
```

```python
plt.figure(figsize=(13,8))
ax=sns.boxplot(x='StName',y='a_co_mean_ln',data=data_new,orient='v')
plt.setp(ax.get_xticklabels(),rotation=45)
plt.show()
```

```python
plt.violinplot(co,quantiles=[0.25,0.5,0.75])
plt.show()
```

```python
plt.figure(figsize=(13,8))
ax=sns.violinplot(x='StName',y='a_co_mean_ln',data=data_new,orient='v')
plt.setp(ax.get_xticklabels(),rotation=45)
plt.show()
```

```python
data_state = data_new[data_new['StName'].isin(['Alabama', 'Alaska','Wiscons'])]
data_state.head()
```

```python
sns.violinplot(x='StName',y='a_co_mean_ln',data=data_state,orient='v')
```

```python
sns.violinplot(x='StName',y='a_co_mean_ln',data=data_state,orient='v',hue='StName')
```
